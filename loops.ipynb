```python
#Basic loops 
#------while Loops------
count = 1
while count <= 5:
    print(f" count is {count }")
    count = count + 1 # count +=1
print ("done with while loop")
```

```python
# Real Life example of while loop
password = "abc123"
attempt = 0
while attempt < 3:
    user_input = input("Enter the password: ")
    if user_input == password:
        print("access granted")
        break
    else:
        print("access denied")
        attempt +=1
        if attempt == 1:
            print("First attempt")
        elif attempt == 2:
            print("Second attempt")
        else:
            print("3 attempts,account locked")
# another way is also possible to do this is by boolean variable.
   
```

```python
#------For Loops------
# range(stop) - from 0 to stop-1
for i in range(5): # i starts from 0 implicitly and goes up to 4
    print(f"i is {i}")

```

```python
#if you want to start from 1 and go up to 5, you can do this
# range(start, stop) - from start to stop-1
for i in range(1,6):
    print(f"i is {i}")
```

```python
#printing even numbers upto 10
for i in range(0, 11, 2): #range (start, stop , increment)
    print(f"i is {i}")
    
```
