```python
#Loop control system
#find even and odd numbers 

numbers = [1,2,3,4,5,6,7,8,10]

for num in numbers:
    if num % 2 == 0:
        print ("even number ", num)
    else:
        print("Odd number", num)
    
```

```python
# skip even number
for i in range (1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#search for item in list
shoppingList = ["milk", "eggs", "bread"]
search = input("what do you want to find? ")
for item in shoppingList:
    if item == search :
        print ("found ", search)
        break
else:
        print("Not found ", search)
        
```

```python
#nested loops

#Draving Pattern 
size = 5
for i in range (size):
    for j in range(size):
        #print("*", end="") #another way to print in same line is 
        print("*", end=" ")
    print() #print a newline after each row
```

```python
#nested loop to print pattern

#Triangle pattern
for i in range (1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
#calculate sum of number user enters
total = 0
count = 0
print("enter numbers to sum, type '0' to stop")
while True:
    num = int(input("Enter a Number: "))
    if num == 0:
        break
    total = total + num  #total +=num also works
    count = count + 1
    print("Current total: ", total)
print()
print("Final Total: ", total)
print('You entered', count, 'numbers')

```

```python
# User choose Pattern
size = int (input("enter size of pattern: "))
print("choose pattern: ")
print("1 square")
print("2 triangle")
print("3 diamond ")

choice = int (input ("enter choice (1-3): "))

if choice == 1:
    for i in range(size):
        for j in range(size):
            print ('*', end = " ")
        print()

elif choice == 2:
    for i in range(1,size + 1):
        for j in range(i):
            print("*", end= " ")
        print()

elif choice == 3:
    for i in range(1, size+1):
        spaces = size - i 
        stars = 2*i-1
        print(" " * spaces + "*" * stars)

    for i in range(size - 1, 0, -1):
        spaces = size -i
        stars = 2* i -1
        print (" " * spaces + "*" * stars)
```
