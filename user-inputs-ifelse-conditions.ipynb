```python
#simple calculator 
print ("Welcome to simple calculator")
print("-"*30) #this print 30 dashes

num1 = int(input("enter your first number:"))
num2 = int(input("enter your second number:"))
sum = num1+num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2 #quoutient
div2 = num1 // num2 #floor division
print(f"sum of {num1} and {num2 } is {sum}")
print(f"subtraction of {num1} and {num2 } is {sub}")
print(f"multiplication of {num1} and {num2} is {mul}")
print(f"division of {num1 } and {num2 } is {div}")
print(f"floor division of { num1} and {num2} is {div2}")
```
